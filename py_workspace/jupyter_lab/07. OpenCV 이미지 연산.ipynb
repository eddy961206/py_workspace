# OpenCV 이미지 연산

### 이미지 크기 및 픽셀 정보 확인

```python
import cv2
image = cv2.imread('2.jpg')
```

```python
# 픽셀 수 및 이미지 크기 확인
print(image.shape) # cv2는 numpy객체를 리턴함. numpy객체의 shape(차원)  - (360, 360, 3). 여기서 3은 색상의 channel 수.
print(image.size)  # numpy객체의 size(원소 갯수)
```

```python
# 이미지 Numpy 객체의 특정 픽셀을 가리킵니다.
px = image[100, 100]
```

```python
# B, G, R 순서로 출력됩니다.
# (단, Gray Scale인 경우에는 B, G, R로 구분되지 않습니다.)
print(px)
```

```python
# R 값만 출력하기
print(px[2])
```

### 특정 범위 픽셀 변경

```python
import cv2
import matplotlib.pyplot as plt
import time
```

```python
image = cv2.imread('2.jpg')
```

```python
start_time = time.time()
for i in range(0, 100):
    for j in range(0, 100):
        image[i, j] = [255, 255, 255]   # 0~99, 0~99px 흰색으로 바꾸기. 이렇게 하면 오래걸림
print("--- %s seconds ---" % (time.time() - start_time))

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
start_time = time.time()
image[0:100, 0:100] = [0, 0, 0]   # numpy슬라이싱 연산 문법. [0:100, 0:100] : y축(행) 0~99, x축(열) 0~99
print("--- %s seconds ---" % (time.time() - start_time))

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

### ROI(Region of Interest: 관심 있는 영역) 추출

```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('2.jpg')

# Numpy Slicing: ROI 처리 가능
roi = image[200:350, 50:200]  # y x 순서임 주의

plt.imshow(cv2.cvtColor(roi, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# ROI 단위로 이미지 복붙하기
image[0:150, 0:150] = roi  # Numpy Slicing

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

### 픽셀별 색상 다루기

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('2.jpg')
image[:, :, 2] = 0  # [ y, x, channel(2) ]  여기서 BGR [2]번째인 Red을 0으로 하겠다(빨간색 없애겠다)

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```
