# OpenCV 이미지 변형

### 이미지 사이즈 변경

#### cv2.resize(image, dsize, fx, fy, interpolation): 이미지의 크기를 조절합니다.  
  
dsize: Manual Size  
fx: 가로 비율  
fy: 세로 비율  
interpolation: 보간법      

INTER_CUBIC: 사이즈를 크게 할 때 주로 사용합니다.  
  
INTER_AREA: 사이즈를 작게 할 때 주로 사용합니다.  
  
보간법은 사이즈가 변할 때 픽셀 사이의 값을 조절하는 방법을 의미합니다.

```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('2.jpg')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))  # 원본 사이즈
plt.show()
print(image.shape)
```

```python
expand = cv2.resize(image, None, fx=2.0, fy=2.0, interpolation=cv2.INTER_CUBIC)  #  세로 가로 2배 확대
plt.imshow(cv2.cvtColor(expand, cv2.COLOR_BGR2RGB))  
plt.show()
print(expand.shape)
```

```python
shrink = cv2.resize(image, None, fx=0.8, fy=0.8, interpolation=cv2.INTER_AREA)   #  세로 가로 0.8배 축소
plt.imshow(cv2.cvtColor(shrink, cv2.COLOR_BGR2RGB))
plt.show()
print(shrink.shape)
```

### 이미지 위치 변경

#### cv2.warpAffine(image, M, dsize): 이미지의 위치를 변경합니다.

M: 변환 행렬  
dsize: Manual Size

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('2.jpg')

# 행과 열 정보만 저장합니다.
height, width = image.shape[:2]   # 여기선 [2] 가 3번째를 가리키는 게 아니네.. 그냥 첫 두번쨰 까지만 가져오기네

M = np.float32([[1, 0, 50], [0, 1, 10]])  # x 50, y 10만큼 평행이동
dst = cv2.warpAffine(image, M, (width, height))

print(M)
plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
plt.show()
```

### 이미지 회전

#### cv2.getRotationMatrix2D(center, angle, scale): 이미지 회전을 위한 변환 행렬을 생성합니다.
  
center: 회전 중심  
angle: 회전 각도  
scale: Scale Factor

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('2.jpg')

# 행과 열 정보만 저장합니다.
height, width = image.shape[:2]

#     cv2.getRotationMatrix2D(          center,               angle,      scale)
M = cv2.getRotationMatrix2D(  (width / 2, height / 2),     90,         0.5)
dst = cv2.warpAffine(image, M, (width, height))

print(M)
plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
plt.show()
```
