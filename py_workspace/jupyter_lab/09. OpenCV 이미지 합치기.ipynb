cv2.add(): Saturation 연산을 수행합니다.

* Saturation: 0보다 작으면 0, 255보다 크면 255로 표현합니다.

np.add(): Modulo 연산을 수행합니다. (잘 안씀)

* Modulo: 256은 0, 257은 1로 표현합니다.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image_1 = cv2.imread('image_1.jpg')  # cv2.imread(''image_1.jpg', cv2.IMREAD_COLOR) 뒤에 cv2.IMREAD_COLOR이게 생략된. default가 이거.
image_2 = cv2.imread('image_2.png')
```

```python
result = cv2.add(image_1, image_2)   # CV에서 제공하는 add 함수 사용
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
result = image_1 + image_2   # numpy에서 제공하는 기본 덧셈 연산 사용
# result = np.add(image_1, image_2)
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.show()
```
