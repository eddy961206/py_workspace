```python
import cv2
import matplotlib.pyplot as plt
```

```python
img_basic = cv2.imread('2.jpg', cv2.IMREAD_COLOR)
plt.imshow(cv2.cvtColor(img_basic, cv2.COLOR_BGR2RGB))   # matplot BGR  RGB    . CV BGR.
plt.show()
```

```python
img_basic = cv2.cvtColor(img_basic, cv2.COLOR_BGR2GRAY)  #BGR  GRAY 
plt.imshow(cv2.cvtColor(img_basic, cv2.COLOR_GRAY2RGB))   
# imshow()  3 channel (RGB)      1 channel  img_basic  3 channel   
plt.show()
```
