```python
import cv2
import matplotlib.pyplot as plt
```

```python
img_basic = cv2.imread('2.jpg', cv2.IMREAD_COLOR)
plt.imshow(cv2.cvtColor(img_basic, cv2.COLOR_BGR2RGB))   # matplot은 BGR이 아닌 RGB로 바꿔줘야 쓸 수 있음. CV는 BGR.
plt.show()
```

```python
img_basic = cv2.cvtColor(img_basic, cv2.COLOR_BGR2GRAY)  #BGR이던 걸 GRAY로 변환
plt.imshow(cv2.cvtColor(img_basic, cv2.COLOR_GRAY2RGB))   
# imshow()메소드 자체가 3 channel 이미지(RGB) 를 인자로 받기 때문에  1 channel로 된 img_basic를 다시 3 channel로 바꿔줘야 나옴 
plt.show()
```
